```python
import pandas as pd
import numpy as np
import re
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer , CountVectorizer
from nltk.stem import WordNetLemmatizer 
```

```python
pos_rev = pd.read_csv('netflix/pos.txt' , sep = '\n' , header= None , encoding='latin-1')
pos_rev['mood'] = 1
pos_rev.rename(columns={0 : 'review'} , inplace = True)
pos_rev
```

```python
neg_rev = pd.read_csv('netflix/negative.txt' , sep = '\n' , header= None , encoding='latin-1')
neg_rev['mood'] = 0
neg_rev.rename(columns={0 : 'review'} , inplace = True)
neg_rev
```

### what we need to do For cleaning

    lower case
    regex (to remove space and single words)
    tokenize
    stop words
    punctuation 
    stemm/lemmetize

```python
pos_rev['review'] = pos_rev.loc[:,'review'].apply(lambda x : x.lower())
pos_rev
```

```python
pos_rev['review'] = pos_rev.loc[:,'review'].apply(lambda x : re.sub(r'@\S+','',x))
pos_rev              # regex used to remove @ or multiple wide space function it will remove @abc 
```

```python
# lematization
sw = stopwords.words('english')
lem = WordNetLemmatizer()
pos_rev['review'] = pos_rev.loc[: , 'review'].apply(lambda x : " ".join([lem.lemmatize(word , pos='v') for word in x.split() if word not in (sw)]))
pos_rev
```

```python

neg_rev['review'] = neg_rev.loc[: , 'review'].apply(lambda x : x.lower())
neg_rev['review'] = neg_rev.loc[: , 'review'].apply(lambda x : re.sub(r'@\S+' , '' , x))
neg_rev['review'] = neg_rev.loc[: , 'review'].apply(lambda x : x.translate(str.maketrans(dict.fromkeys(string.punctuation))))
neg_rev['review'] = neg_rev.loc[: , 'review'].apply(lambda x : " ".join([lem.lemmatize(word , pos='v') for word in x.split() if word not in (sw)]))
neg_rev
```

```python
common_rev = pd.concat([pos_rev,neg_rev]).reset_index(drop=True)
common_rev
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(common_rev['review'].values, common_rev['mood'].values,test_size=0.2, random_state=42)
```

```python
# TF-IDF

vecterizer  = TfidfVectorizer()
train_vectors = vecterizer.fit_transform(xtrain)
test_vectors = vecterizer.transform(xtest)  # we are using transform only because we dont want a insight on data
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report

sv = SVC(kernel = 'linear')
sv.fit(train_vectors,ytrain)
y_pred_svm = sv.predict(test_vectors)
```

```python
report  = classification_report(ytest,y_pred_svm,output_dict= True)
report
```

```python
import joblib
joblib.dump(sv,'svm.pkl')
joblib.dump(vecterizer, 'transform.pkl')
```
